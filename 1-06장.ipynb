# 1주차 6장. 연산자의 이해

```python
# 산술 연산자 예제-코드
weight = 75
height= 1.83

bmi = weight / (height * height)

print("나의 체중은 ", weight , "kg, 키는 ", height , "m 입니다.")
print("계산한 BMI 지수는  ", bmi)
```

```python
# 관계 연산자 예제-코드
weight = 75
height= 1.83

bmi = weight / (height * height)

print("나의 체중은 ", weight , "kg, 키는 ", height , "m 입니다.")
print("계산한 BMI 지수는  ", bmi)

if( bmi > 25 ) :
	print("과체중 입니다.")
```

```python
# 연산자 예제
print(1 + 1)

print(2 ** 3) 

print(2 * 3)

print(8 % 3)

print(8 > 5)

print(8 < 5)
```

```python
# 연산자 우선순위 예제
print(5 + 10 * 15 / 5 - 10)

print((5 + 10) * 15 / 5 - 10)

print((5 + 10 * 15) / 5 - 10)
```

```python
# 퀴즈 1-답안
weight = 75
height= 1.83

bmi = weight / height * height
print(bmi)

bmi = weight / (height * height)
print(bmi)

bmi = weight / height ** 2
print(bmi)

bmi = weight / (height ** 2)
print(bmi)
```
