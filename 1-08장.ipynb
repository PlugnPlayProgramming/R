```python
# if문 예제
count = 10
if count >=10 :
	count = 0
	print("=" * 25)
	print("다시 처음부터 시작합니다")
	print("=" * 25)
```

```python
# if~else문 예제
count = 10
if	count >=10 :
	count = 0
	print("=" * 25)
	print("다시 처음부터 시작합니다")
	print("=" * 25)

else :
	count = count + 1
	print("다음 단계가 실행됩니다")
```

```python
# 조건문 예제
# if 문 예제에서 사용될 변수 
a = 4
b = 5
c = 6

# 기본 비교 if a < b :
print("a is less than b")

if a > b :
  print("a is greater than b")

if a <= b :
  print("a is less than or equal to b") 
else :
  print("a is greater than b")

if a == b :
  print("a is equal to b") 
else :
  print("a and b are not equal")
```

```python
# if~else문 예제2
# brokers commission
amount= int(input("Enter your transaction amount : "))

if amount > 1000000 : 
  comm = amount * 0.5
elif amount > 500000 : 
  comm = amount * 0.7
elif amount > 200000 : 
  comm = amount * 1.0
else :
  comm = 15000
  
print("Transaction amount = ", amount, "and commission = ", comm)
```

```python
# if~else문 예제3
# Fruits lists
fruits = ['apple', 'banana', 'grape', 'orange', 'strawberry'] 
fruit_input = input("좋아하시는 과일을 입력하세요 : ")

if fruit_input in fruits : 
  print("과일이 목록에 있습니다.")
elif fruit_input == "없음" :
  print("좋아하는 과일이 없습니다.")
else :
  print("과일이 목록에 없습니다.")
```
